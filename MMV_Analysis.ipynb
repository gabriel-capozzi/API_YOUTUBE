```python
import pandas as pd
from datetime import datetime
import re
```

```python
videosDf = pd.read_excel('/content/drive/MyDrive/videosMMV.xlsx', index_col =0)
```

```python
videosDf
```

```python
for index, row in videosDf.iterrows():
  videosDf.loc[index,'publishTime'] = row['publishTime'][:-1]
```

```python
for index, row in videosDf.iterrows():
  videosDf.loc[index,'publishDate'] = datetime.strptime(row['publishDate'], '%Y-%m-%d').date()
```

```python
for index, row in videosDf.iterrows():
  videosDf.loc[index,'publishTime'] = datetime.strptime(row['publishTime'],'%H:%M:%S').time()
```

```python
for index, row in videosDf.iterrows():
  videosDf.loc[index,'Year'] =  format(row['publishDate'].year)
```
